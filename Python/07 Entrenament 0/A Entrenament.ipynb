## Exercici 3

Fes una funció (valida_contrasenya) que a partir d'una contrasenya comprovi que compleix els següents requisits:

* Tenir entre 5 i 10 caràcters (ambdós inclosos)
* Tenir almenys dues lletres majúscules
* Tenir almenys tres lletres minúscules
* Tenir almenys dos números

La funció ha de retornar `True` si la contrasenya és vàlida i `False` si no ho és.


Aleshores fes un programa que demani una contrasenya i comprovi si és vàlida o no.

* Si la contrasenya és vàlida el programa mostra: "Contrasenya vàlida"
* Si no ho és mostra: "Contrasenya no vàlida".

````python
# Fes aquí el codi de l'exercici 3
def valida_contrasenya(contrasenya):
    # Comprovar la longitud de la contrasenya (entre 5 i 10 caràcters)
    if len(contrasenya) < 5 or len(contrasenya) > 10:
        return False
    
    # Comptadors per majúscules, minúscules i números
    majuscules = 0
    minuscules = 0
    numeros = 0
    
    # Recórrer cada caràcter de la contrasenya
    for caracter in contrasenya:
        if caracter.isupper():
            majuscules += 1
        elif caracter.islower():
            minuscules += 1
        elif caracter.isdigit():
            numeros += 1
    
    # Comprovar els requisits: almenys 2 majúscules, 3 minúscules i 2 números
    if majuscules >= 2 and minuscules >= 3 and numeros >= 2:
        return True
    else:
        return False

# Programa principal
contrasenya = input("Introdueix una contrasenya: ")

if valida_contrasenya(contrasenya):
    print("Contrasenya vàlida")
else:
    print("Contrasenya no vàlida")
````

## Exercici 4

Fes una funció (calcula_irpf) que calculi l'IRPF que ha de pagar una persona segons els ingressos bruts que té.

Si et cal, fes altres funcions de suport a aquesta funció principal. Sigues organitzat/da!

Aquests són els trams que ha de tenir en compte:
```text
                | Trams IRPF            |Total|
                |-----------------------|-----|
                | De 0 a 12.450€        | 19% |
                | De 12.450€ a 20.200€  | 24% |
                | De 20.200€ a 35.200€  | 30% |
                | De 35.200€ a 60.000€  | 37% |
                | De 60.000€ a 300.000€ | 45% |
                | Més de 300.000€       | 47% |
```
Per exemple:

* Si una persona cobra 10.200€, hauria de pagar el 19% = 1.938€
```text
+   0,19 *   10.200€           = 1.938,0€
                               ----------
                         Total:  1.938,0€
```
* Si una persona cobra 15.000€, hauria de pagar:
```text
    0,24 * (15.000€ - 12.450€) =   606,0€
+   0,19 *  12.450€            = 2.365,5€
                               ----------
                         Total:  2.971,5€
```
* Si una persona cobra 22.000€, hauria de pagar:
```text
    0,3  * (22.000€ - 20.200€) =   540,0€
    0,24 * (20.200€ - 12.450€) = 1.860,0€
+   0,19 *  12.450€            = 2.365,5€
                               ----------
                         Total:  4.765,5€
```
Comprova:
```text
        * 7600 => 1444.0
        * 16000 => 3217.5
        * 26000 => 5965.5
        * 32000 => 7765.5
        * 54500 => 15866.5
        * 64832 => 20075.9
        * 275387 => 114825.65
        * 765915 => 344881.55
```

````python
# Resol aquí l'exercici 4
def calcula_irpf(ingressos):
    # Definir els trams d'IRPF
    trams = [
        (0, 12450, 0.19),
        (12450, 20200, 0.24),
        (20200, 35200, 0.30),
        (35200, 60000, 0.37),
        (60000, 300000, 0.45),
        (300000, float('inf'), 0.47)
    ]
    
    # Funció de suport per calcular l'IRPF d'un tram
    def calcula_tram(base, limit_inferior, limit_superior, percentatge):
        # Retorna el valor a pagar per aquest tram en concret
  
        return (min(base, limit_superior) - limit_inferior) * percentatge

    # Variable per emmagatzemar l'IRPF total
    irpf_total = 0
    
    # Recorrerem cada tram per veure en quin entra l'usuari
    for limit_inferior, limit_superior, percentatge in trams:
        if ingressos > limit_inferior:
            # Calculem l'import a pagar en aquest tram
            irpf_total += calcula_tram(ingressos, limit_inferior, limit_superior, percentatge)
    
    # Retornem l'IRPF total calculat
    return irpf_total


# Programa principal per demanar els ingressos
ingressos = float(input("Introdueix els teus ingressos bruts anuals: "))

# Calcular l'IRPF
irpf = calcula_irpf(ingressos)

# Mostrar el resultat
print(f"L'IRPF a pagar és: {irpf:.2f}€")
````

## Exercici 5

Fes una funció (calcula_tarifa) que determini el cost mensual d'una factura telefònica segons els minuts que una persona ha parlat i el número de missatges (SMS) enviats. 

Considera també el cost fix mensual que inclogui certa quantitat de minuts i SMS gratuïts.

Aquests són els detalls del pla:
```text
Cost fix mensual: 15€ (inclou 200 minuts gratuïts i 50 SMS gratuïts)
Cost per minut addicional: 0,10€
Cost per SMS addicional: 0,05€
```
Exemples:
```text
    * Parlar 150 minuts i enviar 40 SMS: Cost fix de 15€.
    * Parlar 250 minuts i enviar 30 SMS: Cost fix (15€) cost addicional de 50 minuts (5€). Total = 20€.
    * Parlar 220 minuts i enviar 80 SMS: Cost fix (15€), cost addicional de 20 minuts (2€) i 30 SMS (1,5€). Total = 18,5€.
```
Comprova que la funció calcula_tarifa funciona correctament amb els següents valors:
```text
* 148 minuts i 20 SMS => 15€
* 264 minuts i 40 SMS => 21.4€
* 420 minuts i 180 SMS => 43.5€
* 520 minuts i 240 SMS => 56.5€
```

````python
# Resol aquí l'exercici 5
def calcula_tarifa(minuts, sms):
    # Constants del pla
    cost_fix = 15.0  # Cost mensual fix
    minuts_gratuits = 200  # Minuts inclosos
    sms_gratuits = 50  # SMS inclosos
    cost_minut_addicional = 0.10  # Cost per minut addicional
    cost_sms_addicional = 0.05  # Cost per SMS addicional

    # Calcular el cost dels minuts addicionals
    minuts_addicionals = max(0, minuts - minuts_gratuits)  # Minuts addicionals si es superen els 200 gratuïts
    cost_minuts = minuts_addicionals * cost_minut_addicional

    # Calcular el cost dels SMS addicionals
    sms_addicionals = max(0, sms - sms_gratuits)  # SMS addicionals si es superen els 50 gratuïts
    cost_sms = sms_addicionals * cost_sms_addicional

    # Calcular el cost total
    cost_total = cost_fix + cost_minuts + cost_sms

    return cost_total

# Programa principal per provar la funció amb els exemples donats
proves = [
    (148, 20),  # 15€
    (264, 40),  # 21.4€
    (420, 180),  # 43.5€
    (520, 240),  # 56.5€
]

for minuts, sms in proves:
    cost = calcula_tarifa(minuts, sms)
    print(f"{minuts} minuts i {sms} SMS => {cost:.2f}€")
````

## Exercici 6

Escriu una funció (calcula_consum) que determini la factura d'aigua d'un client segons el nombre de litres que ha consumit durant un mes.

Els costos són:
```text
Cost fix mensual: 10€ (inclou els primers 1.000 litres)
Fins a 5.000 litres (inclòs): 0,02€ per litre addicional.
Fins a 10.000 litres (inclòs): 0,03€ per litre addicional.
Més de 10.000 litres: 0,05€ per litre addicional.
```
Per exemple:
* Si un client ha consumit 800 litres, hauria de pagar només el cost fix de 10€.
* Si un client ha consumit 6.000 litres: 
```text
    10€ : Cost fix (1.000 litres)
+   80€ : 0,02€ per 4.000 litres
    30€ : 0,03€ per 1.000 litres
-------            ------
   120€ : Total     6.000 litres
```
* Si et cal, fes altres funcions de suport per a calcula_consum. Sigues organitzat/da!

Exemples per comprovar:
```text
* 800 litres => 10€
* 6.000 litres => 120€
* 9.000 litres => 210€
* 12.000 litres => 340€
```

````python
# Resol aquí l'exercici 6
def calcula_consum(consum_litres):
    # Constants del pla
    cost_fix = 10.0  # Cost fix mensual per als primers 1.000 litres
    limit_inicial = 1000  # Litres inclosos en el cost fix
    cost_litres_tram1 = 0.02  # Cost per litre addicional fins a 5.000 litres
    cost_litres_tram2 = 0.03  # Cost per litre addicional fins a 10.000 litres
    cost_litres_tram3 = 0.05  # Cost per litre addicional més de 10.000 litres

    # Calcular el cost total
    if consum_litres <= limit_inicial:
        return cost_fix  # Si consumeixen 1.000 litres o menys, només el cost fix
    
    # Si consumeixen més de 1.000 litres, calcular els litres addicionals
    litres_restants = consum_litres - limit_inicial
    cost_total = cost_fix
    
    # Calcular el cost per als primers 4.000 litres addicionals (fins a 5.000 litres en total)
    if litres_restants > 4000:
        cost_total += 4000 * cost_litres_tram1
        litres_restants -= 4000
    else:
        cost_total += litres_restants * cost_litres_tram1
        litres_restants = 0

    # Calcular el cost per als següents 5.000 litres addicionals (fins a 10.000 litres en total)
    if litres_restants > 5000:
        cost_total += 5000 * cost_litres_tram2
        litres_restants -= 5000
    else:
        cost_total += litres_restants * cost_litres_tram2
        litres_restants = 0

    # Calcular el cost per als litres restants (més de 10.000 litres)
    cost_total += litres_restants * cost_litres_tram3

    return cost_total

# Programa principal per provar la funció amb els exemples donats
proves = [
    800,   # 10€
    6000,  # 120€
    9000,  # 210€
    12000  # 340€
]

for consum_litres in proves:
    cost = calcula_consum(consum_litres)
    print(f"{consum_litres} litres => {cost:.2f}€")
````

```text
Preu estàndard: 10€
Nens (menors de 12 anys): 5€
Gent gran (65 anys o més): 6€
Dimarts (Dia del client): Totes les entrades tenen un descompte del 20%.
Dijous (Dia familiar): Si un adult compra una entrada, l'entrada del primer nen té un 50% de descompte els altres nens no paguen.
```

Els paràmetres de la funció són:

No feu inputs, feu servir la funció amb els valors d'exemple i un print per cada cas.

Per exemple:
* Un nen que va al cinema un dimarts: 5€ - 20% = 4€.
* Un adult que va al cinema amb un nen un dijous: L'adult paga 10€ i el nen paga 2,50€ (50% de descompte sobre 5€).
* Un adult que va al cinema amb tres nens un dijous: L'adult paga 10€, el primer nen paga 2,50€, els altres res.
* Si et cal, fes altres funcions de suport per a calcula_entrada. Sigues organitzat/da!

Exemples per comprovar:
* Adult un dilluns: 10€
* Nen un dimarts: 4€
* Gent gran un dijous: 6€ (El descompte del dia familiar només s'aplica als nens)
* Adult i nen un dijous: 10€ + 2,50€ = 12,50€

````python
# Resol aquí l'exercici 7
def calcula_entrada(num_adults, num_nens, num_gent_gran, dia):
    # Preus estàndard
    preu_adult = 10.0
    preu_nen = 5.0
    preu_gent_gran = 6.0

    # Descompte del dimarts
    descompte_dimarts = 0.20  # 20%

    # Variable per emmagatzemar el cost total
    cost_total = 0.0

    # Calcular el cost de les entrades per adults
    cost_adults = num_adults * preu_adult
    cost_total += cost_adults

    # Calcular el cost de les entrades per nens
    cost_nens = 0.0
    if dia.lower() == "dijous" and num_adults > 0:
        # Primer nen té descompte del 50% i la resta no paguen
        if num_nens > 0:
            cost_nens += preu_nen * 0.5  # 50% de descompte pel primer nen
        # Els altres nens no paguen res
    else:
        # Si no és dijous, els nens paguen normalment
        cost_nens += num_nens * preu_nen

    # Calcular el cost de les entrades per la gent gran
    cost_gent_gran = num_gent_gran * preu_gent_gran
    cost_total += cost_gent_gran

    # Afegim el cost dels nens calculat abans
    cost_total += cost_nens

    # Si és dimarts, aplicar el descompte del 20% a totes les entrades
    if dia.lower() == "dimarts":
        cost_total *= (1 - descompte_dimarts)

    return cost_total

# Proves amb exemples
print(f"Adult un dilluns: {calcula_entrada(1, 0, 0, 'dilluns'):.2f}€")
print(f"Un nen un dimarts: {calcula_entrada(0, 1, 0, 'dimarts'):.2f}€")
print(f"Gent gran un dijous: {calcula_entrada(0, 0, 1, 'dijous'):.2f}€")
print(f"Adult i nen un dijous: {calcula_entrada(1, 1, 0, 'dijous'):.2f}€")
print(f"Adult i tres nens un dijous: {calcula_entrada(1, 3, 0, 'dijous'):.2f}€")
````

## Exercici 8 

Un petit hotel a la costa vol informatitzar el seu sistema de reserves. 

Necessiten una funció (calcula_reserva) que determini el preu total d'una reserva segons el nombre de nits, el tipus d'habitació i si el client vol esmorzar inclòs.

Les tarifes són:
```text
Habitació estàndard: 50€ per nit.
Habitació amb vistes al mar: 70€ per nit.
Suite amb jacuzzi: 120€ per nit.
Esmorzar: 10€ per dia i persona.
```
La funció hauria d'acceptar els següents paràmetres: 
```text
* nombre de nits
* tipus d'habitació (estàndard, amb vistes o suite)
* el nombre de persones que volen esmorzar.
```
A més, l'hotel ofereix les següents promocions:
```text
* Si reserves més de 5 nits, t'ofereixen un descompte del 10% sobre el total de la reserva (sense comptar l'esmorzar).
* Si reserves una suite amb jacuzzi per més de 3 nits, l'esmorzar és gratuït.
```

Et pot anar bé definir primer les següents funcions:
```text
* calcula_cost_habitacio(tipus_habitacio, nits)
* calcula_cost_esmorzar(nombre_persones, nombre_dies)
```

Per exemple:
```text
* Reserva d'una habitació estàndard per 4 nits sense esmorzar: 50€ * 4 = 200€.
* Reserva d'una habitació amb vistes al mar per 6 nits amb esmorzar per a 2 persones: [(70€ * 6) - 10%] + (10€ * 6 * 2) = 444€.
* Si et cal, fes altres funcions de suport per a calcula_reserva. Sigues organitzat/da!
```

Exemples per comprovar:
```text
* Habitació estàndard, 4 nits, 2 persones amb esmorzar: 280€
* Suite amb jacuzzi, 5 nits, 2 persones amb esmorzar (gratuït per la promoció): 640€
* Habitació amb vistes, 6 nits, 1 persona sense esmorzar: 438€
```

````python
# Resol aquí l'exercici 8
# Funció per calcular el cost de l'habitació segons el tipus i les nits
def calcula_cost_habitacio(tipus_habitacio, nits):
    if tipus_habitacio == "estàndard":
        return 50 * nits
    elif tipus_habitacio == "amb vistes":
        return 70 * nits
    elif tipus_habitacio == "suite":
        return 120 * nits
    else:
        raise ValueError("Tipus d'habitació no vàlid")

# Funció per calcular el cost de l'esmorzar segons el nombre de persones i dies
def calcula_cost_esmorzar(nombre_persones, nombre_dies, tipus_habitacio, nits):
    # Si és una suite i més de 3 nits, l'esmorzar és gratuït
    if tipus_habitacio == "suite" and nits > 3:
        return 0
    return 10 * nombre_persones * nombre_dies

# Funció principal per calcular el total de la reserva
def calcula_reserva(nits, tipus_habitacio, nombre_persones_esmorzar):
    # Calcula el cost de l'habitació
    cost_habitacio = calcula_cost_habitacio(tipus_habitacio, nits)

    # Calcula el cost de l'esmorzar
    cost_esmorzar = calcula_cost_esmorzar(nombre_persones_esmorzar, nits, tipus_habitacio, nits)

    # Aplicar descompte del 10% si la reserva és per més de 5 nits (sense incloure esmorzar)
    if nits > 5:
        cost_habitacio *= 0.90  # Aplicar descompte del 10%

    # Calcula el cost total
    cost_total = cost_habitacio + cost_esmorzar
    return cost_total

# Proves amb exemples
print(f"Habitació estàndard, 4 nits, 2 persones amb esmorzar: {calcula_reserva(4, 'estàndard', 2):.2f}€")
print(f"Suite amb jacuzzi, 5 nits, 2 persones amb esmorzar: {calcula_reserva(5, 'suite', 2):.2f}€")
print(f"Habitació amb vistes, 6 nits, 1 persona sense esmorzar: {calcula_reserva(6, 'amb vistes', 0):.2f}€")
````

````python
# Fes aquí el codi de l'exercici 9
# Funció per calcular el cost del lloc
def calcula_cost_lloc(tipus_lloc):
    if tipus_lloc == "sala estàndard":
        return 100
    elif tipus_lloc == "jardí amb piscina":
        return 200
    elif tipus_lloc == "saló gran amb escenari":
        return 500
    else:
        raise ValueError("Tipus de lloc no vàlid")

# Funció per calcular el cost del menjar
def calcula_cost_menjar(tipus_menjar, num_convidats):
    if tipus_menjar == "bàsic":
        cost_menjar = 15 * num_convidats
    elif tipus_menjar == "premium":
        cost_menjar = 30 * num_convidats
    else:
        raise ValueError("Tipus de menjar no vàlid")

    # Aplicar descompte si hi ha més de 50 convidats
    if num_convidats > 50:
        cost_menjar *= 0.95  # Descompte del 5%

    return cost_menjar

# Funció per calcular el cost de l'entreteniment
def calcula_cost_entreteniment(tipus_entreteniment, tipus_lloc):
    cost_entreteniment = 0
    if "màgia" in tipus_entreteniment:
        cost_entreteniment += 250
    if "música" in tipus_entreteniment:
        cost_entreteniment += 500
        # Descompte si hi ha música en directe i saló gran amb escenari
        if tipus_lloc == "saló gran amb escenari":
            cost_entreteniment -= 100  # Descompte de 100€

    return cost_entreteniment

# Funció principal per calcular el total de la festa
def calcula_festa(tipus_lloc, num_convidats, tipus_menjar, tipus_entreteniment):
    # Calcula el cost del lloc
    cost_lloc = calcula_cost_lloc(tipus_lloc)

    # Calcula el cost del menjar
    cost_menjar = calcula_cost_menjar(tipus_menjar, num_convidats)

    # Calcula el cost de l'entreteniment
    cost_entreteniment = calcula_cost_entreteniment(tipus_entreteniment, tipus_lloc)

    # Calcula el cost total
    cost_total = cost_lloc + cost_menjar + cost_entreteniment
    return cost_total

# Proves amb exemples
print(f"Jardí amb piscina, 40 convidats, menú bàsic, música en directe: {calcula_festa('jardí amb piscina', 40, 'bàsic', ['música']):.2f}€")
print(f"Saló gran amb escenari, 70 convidats, menú premium, música en directe (amb descompte): {calcula_festa('saló gran amb escenari', 70, 'premium', ['música']):.2f}€")
print(f"Sala estàndard, 15 convidats, menú premium, màgia: {calcula_festa('sala estàndard', 15, 'premium', ['màgia']):.2f}€")
````
